```python
import requests
import pandas as pd
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
token="Здесь должен быть токен"
```

```python
group_name="vesti"
offset=0
count=100
all_posts=[]
all_comments=[]
```

# Инфляци

```python
offset=0
while offset<1700:
        req=requests.get('https://api.vk.com/method/wall.search',
               params={
                   'access_token': token,
                   'v': 5.131,
                   'domain': group_name,
                   'owners_only': 1,
                   'query': 'инфляци',
                   'count': count,
                   'offset': offset
                   
               })
        res=req.json()['response']['items']
        offset+=100
        time.sleep(1)
        all_posts.extend(res)
```

```python
df = pd.json_normalize(all_posts)
df=df[["id", "owner_id", "date", "text", "comments.count"]]
df.rename(columns={'date': 'post_time','text':'post_text'}, inplace=True)
df
```

```python
posts_id1=list(df[df['comments.count']!=0]['id'])
```
