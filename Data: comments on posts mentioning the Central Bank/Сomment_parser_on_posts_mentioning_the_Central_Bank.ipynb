```python
import requests
import pandas as pd
import time
import warnings
import re
warnings.filterwarnings('ignore')
```

```python
token="Здесь должен быть токен"
```

```python
group_name="expert_ru"
offset=0
count=100
all_posts=[]
all_comments=[]
```

# банк россии

```python
offset=0
while offset<1700:
        req=requests.get('https://api.vk.com/method/wall.search',
               params={
                   'access_token': token,
                   'v': 5.131,
                   'domain': group_name,
                   'owners_only': 1,
                   'query': 'банк россии',
                   'count': count,
                   'offset': offset
                   
               })
        res=req.json()['response']['items']
        offset+=100
        time.sleep(1)
        all_posts.extend(res)
```

```python
df = pd.json_normalize(all_posts)
df=df[["id", "owner_id", "date", "text", "comments.count"]]
df.rename(columns={'date': 'post_time','text':'post_text'}, inplace=True)
df['bool']=0
for i in range (len(df)):
    s=str(df.iloc[i]['post_text'])
    s=str(s)
    s=s.lower()
    c=bool(re.search(r'банк россии|банка россии|банку россии|банком россии|банке россии|цб|центробанк|центральный банк|центрального банка|центральному банку|центральным банком|центральном банке', s))
    df.loc[i,'bool']=df['bool'][i]+c
df=df[df['bool']==1]
df
```

```python
posts_id1=list(df[df['comments.count']!=0]['id'])
```
